```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'G:\traindata\RFMtest\tmp.xlsx')
df.head()
```

```python
df['trdtime'] = pd.to_datetime(df['trdtime'])
df.info()
```

```python
df['trdtime'].min()
```

```python
R_today = dt.datetime(2020,8,26)
```

```python
df['R_diff'] = (R_today-df['trdtime']).dt.days
df.head()
```

```python
R = df.groupby(by = ['id'])['R_diff']
R.head()
```

```python
R = R.agg([('R_diff','min')])
R.head()
```

```python
F = df.groupby(by = ['id'])['id'].agg([('F_fre','count')])
```

```python
M = df.groupby(by = ['id'])['total'].agg([('M_sum',sum)])
RFM = R.join(F).join(M)
RFM.info()
```

```python
person = RFM.groupby(by = ['F_fre'])['F_fre'].agg([('F_fre','count')])
person.head(20)
```

```python
RFM.describe()
```

```python
RFM["R_score"]=np.where(RFM.R_diff<=17,4,np.where(RFM.R_diff<=25,3,np.where(RFM.R_diff<=33,2,1)))
RFM["F_score"]=np.where(RFM.F_fre<=1,1,np.where(RFM.F_fre<=2,2,np.where(RFM.F_fre<=3,3,4)))
RFM["M_score"]=np.where(RFM.M_sum<=26.52,1,np.where(RFM.M_sum<=44,2,np.where(RFM.M_sum<=84,3,4)))
RFM.head()
```

```python
RFM["RFM"] = RFM["R_score"]*0.3+RFM["F_score"]*0.4+RFM["M_score"]*0.3
RFM.head()
```

```python
bins = RFM.RFM.quantile(q = np.linspace(0,1,num = 9),interpolation = 'nearest')
labels = ['潜在客户','一般发展客户','一般保持客户','一般价值客户','重要挽留客户','重要发展客户','重要保持客户','高价值客户']
RFM['types'] = pd.cut(RFM.RFM,bins = bins,labels = labels,include_lowest = True)
RFM.head()
```

```python
df1 = RFM.reset_index()
df1
df1.to_excel("testYD.xlsx")
```
