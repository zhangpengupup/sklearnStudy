```python
X = train['week']
X = X.values.reshape(-1, 1)
y = train.iloc[:,-2]
y = y.values.reshape(-1, 1)
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=445)
reg = LinearRegression().fit(Xtrain,Ytrain)
reg.score(Xtrain,Ytrain)
reg.score(Xtest,Ytest)
```

```python
y_hat = reg.predict(X)
plt.plot(X,y_hat,'r-')
plt.plot(X,y,'bo')
plt.show
```

```python
reg_coef = reg.coef_     #截距
reg_intercept = reg.intercept_ #斜率
prediction = reg_coef*int('15')+reg_intercept
prediction
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2,include_bias=False)#添加多项式回归，增加拟合程度
Xtrain_poly = poly.fit_transform(Xtrain)#分别对训练集和测试集进行处理
Xtest_poly = poly.fit_transform(Xtest)#
X_poly = poly.fit_transform(X)
modelall = LinearRegression(normalize=True).fit(Xtrain_poly,Ytrain)
j = modelall.score(Xtest_poly,Ytest)
Y_pre = modelall.predict(Xtest_poly)
# plt.plot(Xtest,Y_pre,'ro')
# plt.plot(Xtest,Ytest,'bo')
# plt.show
y_pre = modelall.predict(X_poly)#展示
plt.plot(X,y_pre,'r-')
plt.plot(X,y,'bo')
plt.show
```
