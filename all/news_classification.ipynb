```python
import pandas as pd 
contents,labels = [],[]
with open('G:\\traindata\\cnlag\\cnews\\cnews.test.txt',"r",encoding='utf-8')as f:
    for line in f:
        try:
            label,content = line.strip().split('\t')
            if content:
                contents.append(content)
                labels.append(label)
        except:
            pass
direct = {'content':contents,'label':labels}
df = pd.DataFrame(direct)
df.head()
```

```python
import re 
import jieba

stopwords = [line.strip() for line in open('G:\\traindata\\cnlag\\chineseStopTxt.txt',encoding='UTF-8').readlines()]
def cut_part(sentence):
    sentence = re.sub(r'[^\u4e00-\u9fa5]+','',sentence)
    sentence_depart = jieba.cut(sentence.strip())
    word_list = []
    for word in sentence_depart:
            if word not in stopwords:
                word_list.append(word)
    return word_list

content_word_list = []
for content in df['content']:
    line_seg = cut_part(content)
    content_word_list.append(" ".join(line_seg))
print(len(content_word_list))
print(content_word_list[:3])
```

```python
df['cut_word_list'] = content_word_list
```

```python
df = df[['label','cut_word_list']]
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder

labels = df['label']
text = df['cut_word_list']


X_train,X_test,y_train,y_test = train_test_split(text,labels,random_state = 0,test_size = 0.2)

count_vect = CountVectorizer()#将分割完的词转换为稀疏矩阵，默认mindf=1,也就是最少在一个文档出现过
count_vect
```

```python
X_train_count = count_vect.fit_transform(X_train)
# X_train_count.toarray()
```

```python
tf_transformer = TfidfTransformer().fit(X_train_count)  #词频处理   此处采用的是countvectorizer+TfidfTransformer 处理获得     也可以直接使用TfidfVectorizer直接完成处理
print(tf_transformer)
```
